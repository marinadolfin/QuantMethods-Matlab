Step 3 — Import your Excel data

```python
import pandas as pd  #Load the pandas library for data handling and call it pd.

# pandas is the Python’s way to store, clean, and analyze tabular or time-series data.

data = pd.read_excel("Data.xlsx")  #import the excel file into a pandas DataFrame

data.head()  #display the first rows of data
```

Plot Apple stock  prices against time.

```python
data["Date"] = pd.to_datetime(data["Date"])# check Date is a datetime

date = data["Date"]
stock = data["AAPL"]
```

```python
import matplotlib.pyplot as plt  #Load the Matplotlib plotting library and call it plt for convenience.

plt.figure(figsize=(10,5))
plt.plot(date, stock, label="Price", color="navy")
plt.title("Stock Price Over Time")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.show()
```

Compute Apple logarithmic returns and store them in a variable returns.

```python
data["AAPL"].pct_change() #compute simple returns
```

```python
import numpy as np  #Load the NumPy library for numerical and array operations, and call it np.

returns = np.log(stock / stock.shift(1))

```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


# --- Create the figure with two subplots (2 rows, 1 column) ---
fig, axes = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# --- Top panel: Price ---
axes[0].plot(date, stock, color="navy")
axes[0].set_title(" Price")
axes[0].set_ylabel("Price (USD)")
axes[0].grid(True)

# --- Bottom panel: Log Returns ---
axes[1].plot(date, returns, color="darkred")
axes[1].set_title("Log Returns")
axes[1].set_xlabel("Date")
axes[1].set_ylabel("Log Return")
axes[1].grid(True)

# --- Adjust layout and display ---
plt.tight_layout()
plt.show()
```
